````python
matches=pd.read_csv("matches.csv")
booking=pd.read_csv("booking.csv")
bets=pd.read_csv("bets.csv")
goals=pd.read_csv("goals.csv")
stats=pd.read_csv("stats.csv")
````

````python
na_s=stats.isna().sum()
cols_drop=na_s[na_s>1000].index
stats=stats[stats.columns.difference(cols_drop)]
````

````python
na_s=stats.isna().sum()
rows_dropped=na_s[na_s>10].index
stats[rows_dropped]=stats[rows_dropped].fillna(stats[rows_dropped].mean())
stats.isna().sum()
````

````python
stats=stats.dropna()
````

````python
stats["home_BallPossession"]=stats["home_BallPossession"].apply(lambda x: int(x.strip("%"))/100)
stats["away_BallPossession"]=stats["away_BallPossession"].apply(lambda x: int(x.strip("%"))/100)
````

````python
matches["match_hometeam_name"]=matches["match_hometeam_name"].apply(lambda x:"West Ham" if x=="West Ham (Eng)" else "Manchester Utd" if x=="Manchester United" else "Newcastle" if x=="Newcastle Utd" else x)
matches["match_awayteam_name"]=matches["match_awayteam_name"].apply(lambda x:"Aston Villa" if x=="Aston Villa (Eng)" else "Manchester Utd" if x=="Manchester United" else "Newcastle" if x=="Newcastle Utd" else x)
del matches["match_hometeam_penalty_score"],matches["match_awayteam_penalty_score"]
````

````python
matches.isna().sum()
````

````python
matches.drop(columns=["match_hometeam_halftime_score","match_awayteam_halftime_score","match_hometeam_extra_score","match_awayteam_extra_score"],inplace=True)
matches.dropna(inplace=True)
matches
````

````python
df_merged=pd.merge(matches,stats,how="left",on="match_id")
df_merged.dropna(inplace=True)
df_merged
````

````python
booking=booking[booking.columns.difference(["time","info"])]
booking["home_fault"]=booking["home_fault"].apply(lambda x: 0 if pd.isnull(x) else 1)
booking["away_fault"]=booking["away_fault"].apply(lambda x: 0 if pd.isnull(x) else 1)
booking
````

````python
a=pd.pivot_table(booking,values=["home_fault","away_fault"],columns="card",index="match_id",aggfunc=["mean","sum"])
a.columns=['_'.join(k) for k in a]
a.fillna(0,inplace=True)
a.reset_index(inplace=True)
a
````

````python
df_merged=pd.merge(df_merged,a,how="left",on="match_id").dropna()
df_merged
````

````python
teams_match_ids={}
ids_rem=np.array([])
teams=df_merged["match_hometeam_name"].value_counts().index.values
for team in teams:
    aa=[]
    match_idss=df_merged[(df_merged["match_hometeam_name"]==team)|(df_merged["match_awayteam_name"]==team)].values
    ids_rem=np.append(ids_rem,match_idss[:,2])
    for match in match_idss:
        if match[6]==team:
            aa.append((match[2],"home"))
        elif match[7]==team:
            aa.append((match[2],"away"))
        else:
            aa.append((team,"hata"))
    teams_match_ids[team]=np.array(aa)
ids_rem=np.unique(ids_rem)
# teams_match_ids
````

````python
df_merged[df_merged["match_id"]==ids_rem[0]].values
````

````python
data_cols=['match_{}team_score', '{}_BallPossession',
'{}_BlockedShots', '{}_CornerKicks', '{}_Fouls', '{}_FreeKicks',
'{}_GoalAttempts', '{}_GoalkeeperSaves', '{}_Offsides',
'{}_ShotsoffGoal', '{}_ShotsonGoal', '{}_YellowCards',
'mean_{}_fault_red card', 'mean_{}_fault_yellow card',
'sum_{}_fault_red card', 'sum_{}_fault_yellow card']
````

````python
new_df=[]
from tqdm import tqdm
for m_id in tqdm(ids_rem):
    home_team=df_merged[df_merged["match_id"]==m_id]["match_hometeam_name"].values[0]
    away_team=df_merged[df_merged["match_id"]==m_id]["match_awayteam_name"].values[0]
    game_index=teams_match_ids[home_team][:,0].tolist().index(str(m_id))
    game_index_a=teams_match_ids[away_team][:,0].tolist().index(str(m_id))
    if game_index>5 and game_index_a>5:
        st_home=teams_match_ids[home_team][game_index-5:game_index].copy()
        st_away=teams_match_ids[away_team][game_index_a-5:game_index_a]
        home_stats_tba=[]
        for st_id in st_home:
            needed_cols=[i.format(st_id[1]) for i in data_cols]
            m_stat_home=df_merged[df_merged["match_id"]==int(st_id[0])][needed_cols].values
            home_stats_tba.append(m_stat_home[0])
        home_stats_tba=pd.DataFrame(home_stats_tba,columns=[i.format("home") for i in data_cols])
        new_df_arr_=[m_id]+home_stats_tba.mean().values.tolist()+home_stats_tba.sum().values.tolist()

        away_stats_tba=[]
        for st_id in st_away:
            needed_cols=[i.format(st_id[1]) for i in data_cols]
            m_stat_away=df_merged[df_merged["match_id"]==int(st_id[0])][needed_cols].values
            away_stats_tba.append(m_stat_away[0])
        away_stats_tba=pd.DataFrame(away_stats_tba,columns=[i.format("away") for i in data_cols])
        new_df_arr_=new_df_arr_+away_stats_tba.mean().values.tolist()+away_stats_tba.sum().values.tolist()
        new_df.append(new_df_arr_)
new_df=pd.DataFrame(new_df,columns=["match_id"]+["mean"+i.format("home") for i in data_cols]+["sum"+i.format("home") for i in data_cols]+["mean"+i.format("away") for i in data_cols]+["sum"+i.format("away") for i in data_cols])
new_df
````

````python
new_df=new_df.drop_duplicates().reset_index(drop=True)
new_df.head()
````

````python
new_df=pd.merge(new_df,df_merged[["match_id","match_hometeam_score","match_awayteam_score"]],how="left",on="match_id")
new_df
````

````python
new_df["total_goals"]=new_df["match_hometeam_score"]+new_df["match_awayteam_score"]
new_df["larger_than_2_5"]=new_df["total_goals"].apply(lambda x: 0 if x<2.5 else 1)
new_df.drop(columns=["match_hometeam_score","match_awayteam_score"],inplace=True)
````

````python
new_df[new_df["match_id"].isin(bets.match_id.unique())]
# 1200 mac için yok eklemiyorum
;
````

````python
new_df.reset_index(drop=True,inplace=True)
````

# Feature Engineering

So, basically what I did is:

*   Merge the stats dataframe and the card dataframe with the matches: aggregated them in terms of mean and their summation.
*   Then take the last 5 matches of a team, pull the mean and summation information again from the dataframe.

*   Merged with the matches dataframe again.





````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train_log,y_test_log=train_test_split(new_df.drop(columns=["larger_than_2_5","total_goals"]),new_df["larger_than_2_5"],test_size=0.1)
y_test_reg=new_df["total_goals"].iloc[X_test.index].copy()
y_train_reg=new_df["total_goals"].iloc[X_train.index].copy()
````

# Penalized Regression Approaches (PRA)

````python
print("Lasso Regression For lambda {}:\n\tMean Squared Error:{}\n\tRMSE:{}")
````

````python
classifier_errors={}
regressor_errors={}
````

````python
# Lasso regression
from sklearn.linear_model import Lasso,LogisticRegression
from sklearn import metrics

alphas=[0.5,0.75,1]
reg_err=[]
for alpha in alphas:
    model=Lasso(alpha=alpha)
    model.fit(X_train,y_train_reg)
    score_=model.score(X_train,y_train_reg)
    pred=model.predict(X_test)
    mse_=metrics.mean_squared_error(y_test_reg,pred)
    mae_=metrics.mean_absolute_error(y_test_reg,pred)
    r2_=metrics.r2_score(y_test_reg,pred)
    print("Lasso Regression For lambda {}:\n\tScore:{}\n\tMean Squared Error:{:.4f}\n\tRMSE:{:.4f}\n\tMAE:{:.4f}\n\tR2:{:.4f}".format(alpha,score_,mse_,np.sqrt(mse_),mae_,r2_))
    print("\n")
    reg_err.append([alpha,score_,mse_,np.sqrt(mse_),mae_,r2_])
regressor_errors["Lasso"]=reg_err

# Lasso logistic
alphas=[0.5,0.75,1]
print("-"*45)
clf_err=[]
for alpha in alphas:   
    model=LogisticRegression(penalty="l1",C=alpha,solver="liblinear")
    model.fit(X_train,y_train_log)
    score_=model.score(X_train,y_train_log)
    pred=model.predict(X_test)
    acc=metrics.accuracy_score(y_test_log,pred)
    precision_=metrics.precision_score(y_test_log,pred)
    f1_=metrics.f1_score(y_test_log,pred)
    recall=metrics.recall_score(y_test_log,pred)
    roc_auc=metrics.roc_auc_score(y_test_log,pred)
    print("Logistic Lasso Regression For lambda {}:\n\tScore:{}\n\tAccuracy:{:.4f}\n\tPrecision:{:.4f}\n\tF1 Score:{:.4f}\n\tRecall:{:.4f}\n\tROC-AUC:{:.4f}".format(alpha,score_,acc,precision_,f1_,recall,roc_auc))
    print("\n")
    clf_err.append([alpha,score_,acc,precision_,f1_,recall,roc_auc])
classifier_errors["Lasso"]=clf_err
````

# Decision Tree

````python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.model_selection import ParameterGrid

params_grid={"min_samples_leaf":[1,10,30,50],
             "ccp_alpha":[0,0.5,0.9]}
params_grid=list(ParameterGrid(params_grid))
reg_err=[]

for param_grid in params_grid:
    model=DecisionTreeRegressor(min_samples_leaf=param_grid["min_samples_leaf"],
                                ccp_alpha=param_grid["ccp_alpha"])
    model.fit(X_train,y_train_reg)
    score_=model.score(X_train,y_train_reg)
    pred=model.predict(X_test)
    mse_=metrics.mean_squared_error(y_test_reg,pred)
    mae_=metrics.mean_absolute_error(y_test_reg,pred)
    r2_=metrics.r2_score(y_test_reg,pred)
    print("Decision Tree Regression For min_samples {} and Complexity Parameter {}:\n\tScore:{}\n\tMean Squared Error:{:.4f}\n\tRMSE:{:.4f}\n\tMAE:{:.4f}\n\tR2:{:.4f}".format(param_grid["min_samples_leaf"],param_grid["ccp_alpha"],score_,mse_,np.sqrt(mse_),mae_,r2_))
    print("\n")
    reg_err.append([param_grid,score_,mse_,np.sqrt(mse_),mae_,r2_])
regressor_errors["DT"]=reg_err
params_grid={"min_samples_leaf":[1,10,30,50],
             "ccp_alpha":[0,0.5,0.9]}
params_grid=list(ParameterGrid(params_grid))
clf_err=[]
print("-"*45)
for param_grid in params_grid:
    model=DecisionTreeClassifier(min_samples_leaf=param_grid["min_samples_leaf"],
                                ccp_alpha=param_grid["ccp_alpha"])
    model.fit(X_train,y_train_log)
    score_=model.score(X_train,y_train_log)
    pred=model.predict(X_test)
    acc=metrics.accuracy_score(y_test_log,pred)
    precision_=metrics.precision_score(y_test_log,pred)
    f1_=metrics.f1_score(y_test_log,pred)
    recall=metrics.recall_score(y_test_log,pred)
    roc_auc=metrics.roc_auc_score(y_test_log,pred)
    print("Decision Tree Regression For min_samples {} and Complexity Parameter {}:\n\tScore:{}\n\tAccuracy:{:.4f}\n\tPrecision:{:.4f}\n\tF1 Score:{:.4f}\n\tRecall:{:.4f}\n\tROC-AUC:{:.4f}".format(param_grid["min_samples_leaf"],param_grid["ccp_alpha"],score_,acc,precision_,f1_,recall,roc_auc))
    print("\n")
    clf_err.append([param_grid,score_,acc,precision_,f1_,recall,roc_auc])
classifier_errors["DT"]=clf_err
````

# Random Forest

````python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier

n_feats=[int(2*65/3),20,65]
reg_err=[]
for n_feat in n_feats:
    model=RandomForestRegressor(n_estimators=500,min_samples_leaf=5,max_features=n_feat)
    model.fit(X_train,y_train_reg)
    score_=model.score(X_train,y_train_reg)
    pred=model.predict(X_test)
    mse_=metrics.mean_squared_error(y_test_reg,pred)
    mae_=metrics.mean_absolute_error(y_test_reg,pred)
    r2_=metrics.r2_score(y_test_reg,pred)
    print("Random Forest Regression on {} number of features:\n\tScore:{}\n\tMean Squared Error:{:.4f}\n\tRMSE:{:.4f}\n\tMAE:{:.4f}\n\tR2:{:.4f}".format(n_feat,score_,mse_,np.sqrt(mse_),mae_,r2_))
    print("\n")
    reg_err.append([n_feat,score_,mse_,np.sqrt(mse_),mae_,r2_])
regressor_errors["RF"]=reg_err
n_feats=[int(2*65/3),20,65]
clf_err=[]
print("-"*45)
for n_feat in n_feats:
    model=RandomForestClassifier(n_estimators=500,min_samples_leaf=5,max_features=n_feat)
    model.fit(X_train,y_train_log)
    score_=model.score(X_train,y_train_log)
    pred=model.predict(X_test)
    acc=metrics.accuracy_score(y_test_log,pred)
    precision_=metrics.precision_score(y_test_log,pred)
    f1_=metrics.f1_score(y_test_log,pred)
    recall=metrics.recall_score(y_test_log,pred)
    roc_auc=metrics.roc_auc_score(y_test_log,pred)
    print("Random Forest on {} number of features:\n\tScore:{}\n\tAccuracy:{:.4f}\n\tPrecision:{:.4f}\n\tF1 Score:{:.4f}\n\tRecall:{:.4f}\n\tROC-AUC:{:.4f}".format(n_feat,score_,acc,precision_,f1_,recall,roc_auc))
    print("\n")
    clf_err.append([n_feat,score_,acc,precision_,f1_,recall,roc_auc])
classifier_errors["RF"]=clf_err
````

# Stochastic Gradient Boosting

````python
from sklearn.ensemble import GradientBoostingClassifier,GradientBoostingRegressor

params_grid={"max_depth":[3,6,9,12],
             "learning_rate":[0.1,0.05,0.01],
             "n_estimators":[50,100,150]}
params_grid=list(ParameterGrid(params_grid))
reg_err=[]
for param_grid in params_grid:
    model=GradientBoostingRegressor(n_estimators=param_grid["n_estimators"],learning_rate=param_grid["learning_rate"],max_depth=param_grid["max_depth"])
    model.fit(X_train,y_train_reg)
    score_=model.score(X_train,y_train_reg)
    pred=model.predict(X_test)
    mse_=metrics.mean_squared_error(y_test_reg,pred)
    mae_=metrics.mean_absolute_error(y_test_reg,pred)
    r2_=metrics.r2_score(y_test_reg,pred)
    print("SGB Regression on {} number of features, {} depth, {} learning_rate:\n\tScore:{}\n\tMean Squared Error:{:.4f}\n\tRMSE:{:.4f}\n\tMAE:{:.4f}\n\tR2:{:.4f}".format(param_grid["n_estimators"],param_grid["max_depth"],param_grid["learning_rate"],score_,mse_,np.sqrt(mse_),mae_,r2_))
    print("\n")
    reg_err.append([param_grid,score_,mse_,np.sqrt(mse_),mae_,r2_])
regressor_errors["SGB"]=reg_err

clf_err=[]
params_grid={"max_depth":[3,6,9,12],
             "learning_rate":[0.1,0.05,0.01],
             "n_estimators":[50,100,150]}
params_grid=list(ParameterGrid(params_grid))

print("-"*45)
for param_grid in params_grid:
    model=GradientBoostingClassifier(n_estimators=param_grid["n_estimators"],learning_rate=param_grid["learning_rate"],max_depth=param_grid["max_depth"])
    model.fit(X_train,y_train_log)
    score_=model.score(X_train,y_train_log)
    pred=model.predict(X_test)
    acc=metrics.accuracy_score(y_test_log,pred)
    precision_=metrics.precision_score(y_test_log,pred)
    f1_=metrics.f1_score(y_test_log,pred)
    recall=metrics.recall_score(y_test_log,pred)
    roc_auc=metrics.roc_auc_score(y_test_log,pred)
    print("SGB Classifier on {} number of features, {} depth, {} learning_rate:\n\tScore:{}\n\tAccuracy:{:.4f}\n\tPrecision:{:.4f}\n\tF1 Score:{:.4f}\n\tRecall:{:.4f}\n\tROC-AUC:{:.4f}".format(param_grid["n_estimators"],param_grid["max_depth"],param_grid["learning_rate"],score_,acc,precision_,f1_,recall,roc_auc))
    print("\n")
    clf_err.append([param_grid,score_,acc,precision_,f1_,recall,roc_auc])

classifier_errors["SBG"]=clf_err
````
